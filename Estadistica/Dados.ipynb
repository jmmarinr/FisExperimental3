```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Generate 30 random integers between 1 and 6 (inclusive)
n = 30 # Number of trials
data = np.random.randint(1, 7, size=n)

df = pd.DataFrame(data, columns=['Value'])
media = df.mean()
```

```python
# Counting the results
value_counts = df['Value'].value_counts().sort_values()
```

```python
plt.figure(figsize=(8, 6))  # Set the figure size for better visualization
value_counts.plot(kind='bar', color='red')
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Distribution of Dice trials")
plt.xticks(rotation=0)  # Rotate x-axis labels to prevent overlapping for many categories
plt.tight_layout()
plt.show()
```

```python
df_filtered = df[df['Value'] > 3]

prob = df_filtered.shape[0]/n * 100
print(f'The probability of getting a number above 3 is : {prob:.0f}%')
```
